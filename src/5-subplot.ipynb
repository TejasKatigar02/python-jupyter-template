### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y_line = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# First subplot: Line chart
axs[0, 0].plot(x, y_line, color='blue')
axs[0, 0].set_title('Line Chart: y = x^2')
axs[0, 0].set_xlabel('X values')
axs[0, 0].set_ylabel('Y values')
axs[0, 0].grid(True)

# Second subplot: Bar chart
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title('Bar Chart of Categories and Values')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Third subplot: Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='purple', s=100)
for i in range(len(scatter_x)):
    axs[1, 0].annotate(f'({scatter_x[i]}, {scatter_y[i]})', (scatter_x[i], scatter_y[i]), 
                       textcoords="offset points", xytext=(0,10), ha='center')
axs[1, 0].set_title('Scatter Plot with Annotations')
axs[1, 0].set_xlabel('X values')
axs[1, 0].set_ylabel('Y values')
axs[1, 0].grid(True)

# Fourth subplot: Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
axs[1, 1].set_title('Programming Language Popularity')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
